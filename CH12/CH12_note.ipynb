```python
# Problem 9 (프로그래머스, 문자열 압축 [카카오 2020 공채])
def solution(s):
    answer = []
    L = len(s)
    if L == 1:
        return 1
    
    for i in range(1,L//2+1):
        strs = ''
        bases = s[:i]
        cnt = 1
        for j in range(i,L,i):
            if bases == s[j:j+i]:
                cnt += 1
            else:
                if cnt > 1:
                    strs += str(cnt) + bases
                else:
                    strs += bases
                bases = s[j:j+i]
                cnt = 1
        if cnt > 1:
            strs += str(cnt) + bases
        else:
            strs += bases
        answer.append(len(strs))
    
    return min(answer)
```

```python
# Problem 10 (프로그래머스, 자물쇠와 열쇠 [2020 카카오 신입 공채])
def solution(key, lock):
    N = len(lock)
    M = len(key)
    def rotate_90(key):
        arr = [[0]*M for _ in range(M)]
        for i in range(M):
            for j in range(M):
                arr[j][M-i-1] = key[i][j]
        return arr
    
    lock_E = [[0]*3*N for _ in range(3*N)]
    for i in range(N):
        for j in range(N):
            lock_E[N+i][N+j] = lock[i][j]
    
    def check(lock_E):
        for i in range(N,2*N):
            for j in range(N,2*N):
                if lock_E[i][j] != 1:
                    return False
        return True
    
    for _ in range(4):
        key = rotate_90(key)
        for i in range(N-M+1,2*N):
            for j in range(N-M+1,2*N):
                for k in range(M):
                    for l in range(M):
                        lock_E[i+k][j+l] += key[k][l]
                if check(lock_E):
                    return True
                for k in range(M):
                    for l in range(M):
                        lock_E[i+k][j+l] -= key[k][l]

    return False
```

```python
# Problem 11 (백준 3190)
import sys
inp = sys.stdin.readline
inp = input
N = int(inp())
K = int(inp())

maps = [[0]*N for _ in range(N)]

for _ in range(K):
    i, j = map(int,inp().split())
    maps[i-1][j-1] = 1

rotations = []
L = int(inp())
for _ in range(L):
    logs = inp().split()
    rotations.append((int(logs[0]),logs[1]))
   
# 좌표 : 0 ~ N-1
snake = [[0,0]]
dind = 0
dx = {2:[0,-1], 0:[0,1], 1:[-1,0], 3:[1,0]}
prev_pos = [0,0]
t = 0
end_ind = 0

for r in rotations:
    X, R = r[0], r[1]
    for i in range(t,X):
        prev_pos[0] = snake[0][0]
        prev_pos[1] = snake[0][1]
        snake[0][0] += dx[dind][0]
        snake[0][1] += dx[dind][1]
        t += 1
        if snake[0][0] < 0 or snake[0][0] >= N or snake[0][1] < 0 or snake[0][1] >= N:
            end_ind = 1
            break
        elif snake[0] in snake[1:]:
            end_ind = 1
            break
        if len(snake) > 1:
            for j in range(1,len(snake)):
                snake[j][0], prev_pos[0] = prev_pos[0], snake[j][0]
                snake[j][1], prev_pos[1] = prev_pos[1], snake[j][1]
        if maps[snake[0][0]][snake[0][1]] == 1:
            snake.append([prev_pos[0],prev_pos[1]])
            maps[snake[0][0]][snake[0][1]] = 0

    if end_ind == 1:
        break
    dind = (dind+1)%4 if R == 'L' else (dind-1)%4

while end_ind == 0:
    prev_pos[0] = snake[0][0]
    prev_pos[1] = snake[0][1]
    snake[0][0] += dx[dind][0]
    snake[0][1] += dx[dind][1]
    t += 1
    if snake[0][0] < 0 or snake[0][0] >= N or snake[0][1] < 0 or snake[0][1] >= N:
        break
    elif snake[0] in snake[1:]:
        break
    if len(snake) > 1:
        for j in range(1,len(snake)):
            snake[j][0], prev_pos[0] = prev_pos[0], snake[j][0]
            snake[j][1], prev_pos[1] = prev_pos[1], snake[j][1]
    if maps[snake[0][0]][snake[0][1]] == 1:
        snake.append([prev_pos[0],prev_pos[1]])
        maps[snake[0][0]][snake[0][1]] = 0
    if snake[0] in snake[1:]:
        break
print(t)
```

```python
# Problem 13
import sys
from itertools import combinations
inp = sys.stdin.readline
inp = input

N, M = map(int,inp().split())

house = []
chicken = []

for i in range(N):
    tmp = list(map(int,inp().split()))
    for j in range(N):
        if tmp[j] == 1:
            house.append((i+1,j+1))
        elif tmp[j] == 2:
            chicken.append((i+1,j+1))
            


sum_list = []
for Mchics in combinations(chicken, M):
    sum_dist = 0
    for h in house:
        d = 1000
        for c in Mchics:
            d = min(d,abs(h[0]-c[0])+abs(h[1]-c[1]))
        sum_dist += d
    sum_list.append(sum_dist)
print(min(sum_list))
```
