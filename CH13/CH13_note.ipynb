```python
# Problem 15 (백준 18352)
from collections import deque
def BFS(graph, start, visited, K):
    q = deque([(start,0)])
    arr = []
    while q:
        now, d = q.popleft()
        if visited[now] != 0:
            continue
        visited[now] = 1
        if d == K:
            arr.append(now)
        for node in graph[now]:
            if visited[node] == 0:
                q.append((node,d+1))
    return arr

import sys
inp = sys.stdin.readline

N, M, K, X = map(int,inp().split())

visited = [0] * (N+1)
graph = [[] for _ in range(N+1)]

for i in range(M):
    a, b = map(int,inp().split())
    graph[a].append(b)
    
arr = BFS(graph,X,visited,K)

if len(arr) == 0:
    print(-1)
else:
    arr.sort()
    for node in arr:
        print(node)
```

```python
# Problem 16 (백준 14502)
from collections import deque, Counter
from itertools import combinations
import copy

N, M = map(int,input().split())

lab_map = []

for i in range(N):
    lab_map.append(list(map(int,input().split())))
    
virus = []
empty = []
for i in range(N):
    for j in range(M):
        if lab_map[i][j] == 2:
            virus.append((i,j))
        elif lab_map[i][j] == 0:
            empty.append((i,j))

n0 = len(empty)            
dX = [(1,0),(0,1),(-1,0),(0,-1)] 
answer = 0


for E in combinations(empty,3):
    map_copy = copy.deepcopy(lab_map)
    # 초기 0의 개수는 벽 3개를 만드므로, -3
    # BFS에서 바이러스 시작 위치에서 -1을 해주므로, 초기 바이러스 개수만큼 더함
    num = n0 - 3 + len(virus)
    for e in E:
        map_copy[e[0]][e[1]] = 1
    # BFS
    for v in virus:
        q = deque([v])
        map_copy[v[0]][v[1]] = 0
        while q:
            x, y = q.popleft()
            if map_copy[x][y] > 0:
                continue
            map_copy[x][y] = 2
            num -= 1
            for dx, dy in dX:
                nx, ny = x + dx, y + dy
                if nx < 0 or nx >= N or ny < 0 or ny >= M:
                    continue
                elif map_copy[nx][ny] == 0:
                    q.append((nx,ny))
    answer = max(num,answer)
print(answer)
```

```python
# Problem 18 (프로그래머스, 60058 [카카오 2020 신입공채])
def make_uv(s):
    c1, c2 = 0, 0
    ind = 0
    for i in range(len(s)):
        if s[i] == '(':
            c1 += 1
        else:
            c2 += 1
        if c1 == c2:
            return s[:i+1], s[i+1:]

def is_correct(strs):
    stack = []
    for s in strs:
        if s == '(':
            stack.append(s)
        elif stack:
            stack.pop()
        else:
            return False
    return True if not stack else False

def solution(p):
    answer = ''
    # step 1.
    if p == '':
        return ''
    
    # step 2.
    u, v = make_uv(p)
    
    # step 3.
    if is_correct(u):
        return u + solution(v)
    
    else:
        answer = '(' + solution(v) + ')'
        for i in range(1,len(u)-1):
            if u[i] == '(':
                answer += ')'
            else:
                answer += '('
        return answer
```

```python
#Problem 20 (백준 18428)
# 1. 재귀적으로 구현
import sys, copy
inp = sys.stdin.readline
inp = input
N = int(inp())

obj_to_num = {'X':0,'S':1,'T':2}
teachers = []
empty = []
maps = [[0]*N for _ in range(N)]
dX = [(1,0),(0,1),(-1,0),(0,-1)]
for i in range(N):
    row = inp().split()
    for j in range(N):
        s = row[j]
        maps[i][j] = obj_to_num[s]
        if s == 'X':
            empty.append((i,j))
        elif s == 'T':
            teachers.append((i,j))

def DFS(map_cpy, x, y, direction):
    if x < 0 or x >= N or y < 0 or y >= N:
        return False
    elif map_cpy[x][y] == 1:
        return True
    elif map_cpy[x][y] == 3:
        return False

    
    if direction != -1:
        dx, dy = dX[direction]
        return DFS(map_cpy,x+dx,y+dy,direction)
    
    for i in range(4):
        dx, dy = dX[i]
        if DFS(map_cpy,x+dx,y+dy,i) == True:
            return True
    return False

success = False
for O in combinations(empty,3):
    map_cpy = copy.deepcopy(maps)
    for x, y in O:
        map_cpy[x][y] = 3
    for xt, yt in teachers:
        if DFS(map_cpy,xt,yt,-1):
            success = False
            break
        else:
            success = True
    if success:
        break

if success:
    print('YES')
else:
    print('NO')
```

```python
#Problem 20 (백준 18428)
# 2. 반복문으로 구현
import sys
from itertools import combinations
inp = sys.stdin.readline
inp = input
N = int(inp())

teachers = []
empty = []
maps = [[0]*N for _ in range(N)]

for i in range(N):
    r = inp().split()
    for j in range(N):
        s = r[j]
        maps[i][j] = s
        if s == 'X':
            empty.append((i,j))
        elif s == 'T':
            teachers.append((i,j))

def search_student(maps, x, y):
    # 왼쪽 감시
    xp, yp = x, y
    while y >= 0:
        if maps[x][y] == 'S':
            return True
        elif maps[x][y] == 'O':
            break
        y -= 1
    x, y = xp, yp
    # 오른쪽 감시
    while y < N:
        if maps[x][y] == 'S':
            return True
        elif maps[x][y] == 'O':
            break
        y += 1
    # 위쪽 감시
    x, y = xp, yp
    while x >= 0:
        if maps[x][y] == 'S':
            return True
        elif maps[x][y] == 'O':
            break
        x -= 1
    # 아래족 감시
    x, y = xp, yp
    while x < N:
        if maps[x][y] == 'S':
            return True
        elif maps[x][y] == 'O':
            break
        x += 1
    return False
        

success = False
for O in combinations(empty,3):
    for x, y in O:
        maps[x][y] = 'O'
    for xt, yt in teachers:
        if search_student(maps,xt,yt):
            success = False
            break
        else:
            success = True
    if success:
        break
    for x, y in O:
        maps[x][y] = 'X'

if success:
    print('YES')
else:
    print('NO')
```

```python
# Problem 21 (백준 16234)
import sys
from collections import deque
inp = sys.stdin.readline
inp = input

N, L, R = map(int,inp().split())

A = []

for _ in range(N):
    A.append(list(map(int,inp().split())))
    
answer = 0

dX = [(1,0),(0,1),(-1,0),(0,-1)]

def BFS(A,start,visited,count):
    q = deque()
    q.append((start[0],start[1]))
    num = 0
    group = []
    
    while q:
        x, y = q.popleft()
        if visited[x][y] != count:
            continue
        group.append((x,y))
        num += A[x][y]
        visited[x][y] = count + 1
        for dx, dy in dX:
            nx, ny = x + dx, y + dy
            if nx < 0 or nx >= N or ny < 0 or ny >= N:
                continue
            elif visited[nx][ny] == count and (L <= abs(A[x][y] - A[nx][ny]) <= R):
                q.append((nx,ny))
    if len(group) > 1:
        avg_num = num // len(group)
        for x, y in group:
            A[x][y] = avg_num
        return 1
    return 0

visited = [[0]*N for _ in range(N)]

while True:
    action = 0
    for i in range(N):
        for j in range(N):
            if visited[i][j] == answer:
                action += BFS(A, (i,j), visited,answer)
    if action == 0:
        break
        
    answer += 1

                        
print(answer)
```

```python
# Problem 22 (프로그래머스, 60063 [카카오 2020 신입 공채])
from collections import deque

def next_positions(pos, maps):
    next_pos = []
    pos1, pos2 = list(pos)
    dX = [1,0,-1,0]
    dY = [0,1,0,-1]
    
    for dx, dy in zip(dX,dY):
        np1, np2 = (pos1[0]+dx, pos1[1]+dy), (pos2[0]+dx, pos2[1]+dy)
        if maps[np1[0]][np1[1]] == 0 and maps[np2[0]][np2[1]] == 0:
            next_pos.append({np1,np2})
            
    if pos1[0] == pos2[0]:
        for i in [-1,1]:
            if maps[pos1[0]+i][pos1[1]] == 0 and maps[pos2[0]+i][pos2[1]] == 0:
                next_pos.append({pos1,(pos1[0]+i,pos1[1])})
                next_pos.append({pos2,(pos2[0]+i,pos2[1])})
    elif pos1[1] == pos2[1]:
        for i in [-1,1]:
            if maps[pos1[0]][pos1[1]+i] == 0 and maps[pos2[0]][pos2[1]+i] == 0:
                next_pos.append({pos1,(pos1[0],pos1[1]+i)})
                next_pos.append({pos2,(pos2[0],pos2[1]+i)})
    return next_pos


def solution(board):
    answer = 0
    N = len(board)
    board_e = [[1]*(N+2) for _ in range(N+2)]
    for i in range(N):
        for j in range(N):
            board_e[i+1][j+1] = board[i][j]
    
    q = deque()
    visited = []
    pos = {(1,1),(1,2)}
    q.append((pos,0))
    visited.append(pos)
    
    while q:
        pos, t = q.popleft()
        if (N,N) in pos:
            return t

        for np in next_positions(pos,board_e):
            if np not in visited:
                q.append((np,t+1))
                visited.append(np)
    
    return 0
```
