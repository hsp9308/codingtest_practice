```python
# Problem 37 (백준 11404)
import sys
inp = sys.stdin.readline
inp = input

n = int(inp())
m = int(inp())
INF = 987654321

cost_matrix = [[INF]*(n+1) for _ in range(n+1)]

for i in range(1,n+1):
    cost_matrix[i][i] = 0
    
for _ in range(m):
    a, b, c = map(int,inp().split())
    cost_matrix[a][b] = min(cost_matrix[a][b],c)

for k in range(1,n+1):
    for i in range(1,n+1):
        for j in range(1,n+1):
            cost_matrix[i][j] = min(cost_matrix[i][j],cost_matrix[i][k]+cost_matrix[k][j])

print(cost_matrix)
for i in range(1,n+1):
    for j in range(1,n+1):
        val = cost_matrix[i][j]
        if val == INF:
            print(0,end=' ')
        else:
            print(val,end=' ')
    print()
```

```python
# Problem 38 (K 대회?)
# 플로이드-머셜 사용하지 않고 풀기 (이렇게 정답이 나올지 모르겠음)
n, m = map(int,input().split())

left, right = [set() for i in range(n+1)], [set() for i in range(n+1)]

for i in range(m):
    a, b = map(int,input().split())
    left[b].add(a)
    right[a].add(b)
    
# l : lower, u : upper
for i in range(1,n+1):
    for lower in left[i]:
        left[i] = left[i].union(left[lower])
    for upper in right[i]:
        right[i] = right[i].union(right[upper])
count = 0    
for i in range(1,n+1):
    print(left[i],right[i])
    if len(left[i]) + len(right[i]) >= n-1:
        count += 1
print(count)
```

```python
# Problem 38 -> 플로이드-머셜
import heapq
n, m = map(int,input().split())
INF = 987654321

adj_mat = [[INF]*(n) for _ in range(n)]

for i in range(n):
    adj_mat[i][i] = 0

for i in range(m):
    a, b = map(int,input().split())
    adj_mat[a-1][b-1] = 1

# Floyd-Marshall
for k in range(n):
    for i in range(n):
        for j in range(n):
            adj_mat[i][j] = min(adj_mat[i][j],adj_mat[i][k]+adj_mat[k][j])
answer = 0
for i in range(n):
    count = 0
    for j in range(n):
        if adj_mat[i][j] != INF or adj_mat[j][i] != INF:
            count += 1
    if count == n:
        answer += 1
print(answer)
```

```python
# Problem 39 (ACM-ICPC)

import heapq
inp = input

T = int(inp())
INF = 987654321
answer = []

def dijkstra(graph,distance,start):
    q = []
    x, y = start
    heapq.heappush(q,(graph[x][y],x,y))
    L = len(graph)
    
    dx = [0,0,1,-1]
    dy = [1,-1,0,0]
    
    while q:
        c, x, y = heapq.heappop(q)
        if distance[x][y] < c:
            continue
        for i in range(4):
            nx, ny = x + dx[i], y + dy[i]
            if nx < 0 or nx >= L or ny < 0 or ny >= L:
                continue
            cost = c + graph[nx][ny]
            if cost < distance[nx][ny]:
                distance[nx][ny] = cost
                heapq.heappush(q,(cost,nx,ny))

for _ in range(T):
    N = int(inp())
    graph = []
    for _ in range(N):
        graph.append(list(map(int,inp().split())))
    distance = [[INF]*N for _ in range(N)]
    dijkstra(graph,distance,(0,0))
    answer.append(distance[N-1][N-1])

for val in answer:
    print(val)
```

```python
# Problem 40 (USACO)
# 헛간 번호는 -1 처리했음
import heapq
inp = input
N, M = map(int,inp().split())

graph = [[] for _ in range(N)]

for _ in range(M):
    a, b = map(int,inp().split())
    graph[a-1].append(b-1)
    graph[b-1].append(a-1)
    
INF = 987654321

distance = [INF]*N
distance[0] = 0
# Dijkstra
q = []

heapq.heappush(q,(0,0))

answer = 0
max_dist = 0

while q:
    d, now = heapq.heappop(q)
    if distance[now] < d:
        continue
    for node in graph[now]:
        cost = d + 1
        if cost < distance[node]:
            distance[node] = cost
            heapq.heappush(q,(cost,node))
            if cost > max_dist:
                answer = node
                max_dist = cost
            elif cost == max_dist:
                answer = min(answer,node)
                
print(answer+1, max_dist, distance.count(max_dist))
```
